```python
import pandas as pd
import numpy as np
```

```python
dataset=pd.read_csv("Social_Network_Ads.csv")
dataset
```

```python
dataset=pd.get_dummies(dataset,dtype=int,drop_first=True)
dataset
```

```python
dataset=dataset.drop("User ID",axis=1)
dataset
```

```python
dataset["Purchased"].value_counts()
```

```python
dataset.columns
```

```python
independent=dataset[['Age', 'EstimatedSalary','Gender_Male']]
independent
```

```python
dependent=dataset[['Purchased']]
dependent
```

```python
from sklearn.model_selection import train_test_split
X_train,X_test,Y_train,Y_test=train_test_split(independent,dependent, test_size=0.30, random_state=0)
```

```python
from sklearn.preprocessing import StandardScaler
sc=StandardScaler()
X_train=sc.fit_transform(X_train)
X_test=sc.transform(X_test)
```

```python
from sklearn.svm import SVC
classifier=SVC (kernel= 'rbf', random_state=0) #WITHOUT GRID METHOD
classifier.fit(X_train,Y_train)
```

```python
y_pred=classifier.predict(X_test) #old method
```

```python
from sklearn.svm import SVC
from sklearn.model_selection import GridSearchCV
param_grid= {'kernel': ['linear', 'poly', 'rbf', 'sigmoid'],
                  'C': [ 10,100,1000,2000,3000], 'gamma' :['scale', 'auto']}

grid=GridSearchCV(SVC(probability=True),param_grid,refit= True,verbose=3,n_jobs=-1,scoring= 'f1_weighted')
grid.fit(X_train,Y_train)
                   
```

```python
from sklearn.ensemble import RandomForestClassifier
from sklearn.model_selection import GridSearchCV
param_grid= {'criterion' :['gini','entropy'],
            'n_estimators': [10,100],'max_features':['sqrt','log2','none']}

grid=GridSearchCV(RandomForestClassifier(),param_grid,refit=True,verbose=3,n_jobs=-1,scoring='f1_weighted')
grid.fit(X_train,Y_train)
```

```python
from sklearn.tree import DecisionTreeClassifier
from sklearn.model_selection import GridSearchCV
param_grid= {'criterion' :['gini','entropy'],
            'max_features':['sqrt','log2','none'],'splitter' :['best','random']}

grid=GridSearchCV(DecisionTreeClassifier(),param_grid,refit=True,verbose=3,n_jobs=-1,scoring='f1_weighted')
grid.fit(X_train,Y_train)
```

```python
re=grid.cv_results_
grid_predictions=grid.predict(X_test)
```

```python
from sklearn.metrics import confusion_matrix
cm=confusion_matrix(Y_test,grid_predictions)
```

```python
from sklearn.metrics import classification_report
clf_rep=classification_report(Y_test,grid_predictions)
```

```python
print(clf_rep)
```

```python
print(cm)
```

```python
from sklearn.metrics import roc_auc_score
roc_auc_score(Y_test,grid.predict_proba(X_test)[:,1])
```
